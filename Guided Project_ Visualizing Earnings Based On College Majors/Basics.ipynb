# **Introduction**
We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

   - ``Rank`` - Rank by median earnings (the dataset is ordered by this column).
   - ``Major_code`` - Major code.
   - ``Major`` - Major description.
   - ``Major_category`` - Category of major.
   - ``Total`` - Total number of people with major.
   - ``Sample_size`` - Sample size (unweighted) of full-time.
   - ``Men`` - Male graduates.
   - ``Women`` - Female graduates.
   - ``ShareWomen`` - Women as share of total.
   - ``Employed`` - Number employed.
   - ``Median`` - Median salary of full-time, year-round workers.
   - ``Low_wage_jobs`` - Number in low-wage service jobs.
   - ``Full_time`` - Number employed 35 hours or more.
   - ``Part_time`` - Number employed less than 35 hours.

we are going to use scatter plots and histograms to answer the following questions:
   - Do students in more popular majors make more money?
   - Do students that majored in subjects that were majority female make more money?
   - Is there any link between the number of full-time employees and median salary?
   - What's the most common median salary range?

````python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
````

## Do students in more popular majors make more money?
   To answer this question we are going to create a scatter plot between ```Sample_Size``` (as an indicator for major popularity) and ```Median``` columns.

````python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
````

We can see that even in a very unpopular majors a student can make more money than another student in a more popular major. But what we saw tell now doesn't tell us the whole picture because we don't know the percentage of students who can actually find the job that pays them that much or find a job at all, so next we are going to plot ```Sample_size``` against ```Unemployment_rate``` to see the chances of the students of getting a job

````python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
````

````python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', xlim=(0,500))
````

**Conclusion:** we can say that in a less-popular majors students who find jobs which pays more than a job at a more-popular majors but in popular majors a student are more likely to find a job and the median salary is most likely to be higher than most of the less-popular majors, so the answer to this question would be **YES**.

##    Do students that majored in subjects that were majority female make more money?
To answer this question we are going to compare the ```ShareWomen``` with the ```Unemployment_rate``` and ```Median```

````python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
````

````python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
````

````python
recent_grads.plot(x='Women', y='Median', kind='scatter')
````

````python
recent_grads.plot(x='Men', y='Median', kind='scatter')
````

**Conclusion:** there's no relationship between the number of women in a major and the unemployment rate but we can see that the majors with high women share tends to have less salary which makes the answer to our question is **NO**.
