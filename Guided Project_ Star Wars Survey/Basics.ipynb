# Turn yes/no columns into boolean

```python
yes_no_cols=['Have you seen any of the 6 films in the Star Wars franchise?',
 'Do you consider yourself to be a fan of the Star Wars film franchise?', 'Are you familiar with the Expanded Universe?',
 'Do you consider yourself to be a fan of the Expanded Universe?Â\x8cÃ¦',
 'Do you consider yourself to be a fan of the Star Trek franchise?',]
```

```python
for i in yes_no_cols:
    print(star_wars[i].value_counts())
    print()
```

```python
yes_no_map = {'Yes':True,'No':False}
for i in yes_no_cols:
    star_wars[i]=star_wars[i].map(yes_no_map)
```

# Transform the seen movies into boolean

```python
seen_movies = [ 'Which of the following Star Wars films have you seen? Please select all that apply.',
 'Unnamed: 4',
 'Unnamed: 5',
 'Unnamed: 6',
 'Unnamed: 7',
 'Unnamed: 8',]
```

```python
rename_seen={}
ind=0
for i in seen_movies:
    rename_seen[i] = 'seen_movie_'+str(ind)
    ind+=1
star_wars = star_wars.rename(columns=rename_seen)
seen_movies = rename_seen.values()
```

```python
for i in seen_movies:
    print(star_wars[i].value_counts())
    print()
```

```python
for i in seen_movies:
    star_wars.loc[star_wars[i].notna(),i] = True
    star_wars.loc[star_wars[i].isna(),i] = False
```

```python
for i in seen_movies:
    print(star_wars[i].value_counts())
    print()
```

# Transform ranks into float

```python
user_rank = [ 'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.',
 'Unnamed: 10',
 'Unnamed: 11',
 'Unnamed: 12',
 'Unnamed: 13',
 'Unnamed: 14']
```

```python
star_wars[user_rank].info()
```

```python
rename_rank={}
ind=0
for i in user_rank:
    rename_rank[i] = 'rank_movie_'+str(ind)
    ind+=1
star_wars = star_wars.rename(columns=rename_rank)
user_rank = list(rename_rank.values())
```

```python
user_rank
```

```python
star_wars[user_rank] = star_wars[user_rank].astype(float)
```

```python
star_wars[user_rank].describe()
```

# create a rank mapping to the characters

```python
chr_fam=[  'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.',
 'Unnamed: 16',
 'Unnamed: 17',
 'Unnamed: 18',
 'Unnamed: 19',
 'Unnamed: 20',
 'Unnamed: 21',
 'Unnamed: 22',
 'Unnamed: 23',
 'Unnamed: 24',
 'Unnamed: 25',
 'Unnamed: 26',
 'Unnamed: 27',
 'Unnamed: 28']
```

```python
star_wars[chr_fam].head()
```

```python
for i in chr_fam:
    print(list(star_wars[i].unique()))
    print()
```

```python
rename_chr_rank={}
ind=0
for i in chr_fam:
    rename_chr_rank[i] = 'chr_rank_'+str(ind)
    ind+=1
star_wars = star_wars.rename(columns=rename_chr_rank)
chr_fam = rename_chr_rank.values()
```

```python
import numpy as np
rank_dict = {'Very favorably':5, np.nan:0
             , 'Unfamiliar (N/A)':0
             , 'Somewhat favorably':4
             , 'Very unfavorably':1
             , 'Neither favorably nor unfavorably (neutral)':3
             , 'Somewhat unfavorably':2}
```

```python
for i in chr_fam:
    star_wars[i]=star_wars[i].map(rank_dict)
```
